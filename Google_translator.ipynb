```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, src_lang="fr", tgt_lang="en"):
    model_name = f"Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    inputs = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**inputs)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translated_text

# === Run Translator ===
if __name__ == "__main__":
    # Get user input
    text = input("📝 Enter the text to translate: ")
    src_lang = input("🌐 Enter source language code (e.g., fr, de, hi): ")
    tgt_lang = input("🌐 Enter target language code (e.g., en, fr, hi): ")

    # Translate and show output
    translated = translate_text(text, src_lang, tgt_lang)
    print("\n📄 Translated Text:", translated)
```
